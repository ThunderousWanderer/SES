# Import UCS Satellite Database

```python
import pandas as pd
from google_images_download import google_images_download

satellites = pd.read_excel('C:/Users/Yannick/Desktop/SES/UCS_Satellite_Database_4-1-2019.xlsx')
```

# generate list of search terms for MEO & GEO satellites

```python
satellites['searchTerms'] = satellites['Name of Satellite, Alternate Names'].str.replace("'|,|-|'|/|\*|ä|ö|ü|é|à|è", ' ') + ' ' + satellites['Operator/Owner'].str.replace("'|,|-|'|/|\*|ä|ö|ü|é|à|è", ' ')
#satellites['searchTerms'] = satellites['Name of Satellite, Alternate Names'].str.replace("!@#$%^&*()[]{};:,./<>?\|`~-=_+öäüéàè", " ")  + ' ' + satellites['Operator/Owner'].str.replace("!@#$%^&*()[]{};:,./<>?\|`~-=_+öäüéàè", " ")
satellites.loc[satellites['Class of Orbit'].isin(['GEO','MEO'])]
searchTerms = satellites.loc[satellites['Class of Orbit'].isin(['GEO','MEO'])].searchTerms.tolist()

seperator = ', '
print(seperator.join(searchTerms))
```

# Download for each satellite the 50 first images from Google Image Search

```python
response = google_images_download.googleimagesdownload()

arguments = {"keywords":seperator.join(searchTerms),"output_directory":"C:\\Users\\Yannick\\Desktop\\SES\\picRepo","limit":50,"print_urls":True}
paths = response.download(arguments)
print(paths)
```
